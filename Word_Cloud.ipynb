```
#Importing libraries
import wordcloud
from wordcloud import WordCloud, STOPWORDS
import wikipedia
import re
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Searching wikipedia for a topic and extracting text 
keyword= input("Enter topic for wikipedia search : ")
data_wiki = wikipedia.page(keyword)
text_data = data_wiki.content
# Cleanling the extracted text
text_data = re.sub(r'==.*?==+', '', text_data)
text_data = text_data.replace('\n', '')
print(text_data)
```

```
# Generating our word cloud plot
wordcloud = WordCloud(width = 3000, height = 2000,random_state=1,
                      collocations=False, colormap='gist_rainbow_r',
                      stopwords = STOPWORDS).generate(text_data)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud) 
plt.axis("off");
plt.savefig('data_cloud.jpg', format=None) #saves word cloud as image
```

# Making Word Cloud Attractive

```
# Creating a mask of an image
mask = np.array(Image.open('bulb.jpg'))
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1,
                      background_color='black', colormap='spring', 
                      collocations=False, stopwords = STOPWORDS,
                      mask=mask).generate(text_data)
plt.figure(figsize=(15, 12))
plt.imshow(wordcloud) 
plt.axis("off");
plt.savefig('cloud_design.jpg', format=None) #saves word cloud as image
```
